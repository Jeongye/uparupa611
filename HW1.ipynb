```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.init as init
import time
```

```python
num_data = 1000
num_epoch = 500

x = init.uniform_(torch.Tensor(num_data,1),-5,5)
noise = init.normal_(torch.FloatTensor(num_data,1),std=2)

y1 = x**4+x**2
y1_noise = (x+noise)**4+(x+noise)**2

y2 = x**2+x
y2_noise = (x+noise)**2+(x+noise)
```

```python
model = nn.Linear(1,1)

loss_function = nn.MSELoss()
#손실함수로 L2 loss fuction, 즉 MSE를 사용하므로

learning_rate = 0.01

optimizer = optim.SGD(model.parameters(),lr=0.01)
```

```python
label1 = y1_noise

begin = time.time()

for i in range(num_epoch):
    
    optimizer.zero_grad()
    output = model(x)
    
    loss = loss_function(output,label1)
    
    loss.backward()
    
    optimizer.step()
    
    if i% 10 == 0:
        print(loss.data)
    
print(time.time()-begin)
```

```python
label2 = y2_noise

begin = time.time()

for i in range(num_epoch):
    
    optimizer.zero_grad()
    output = model(x)
    
    loss = loss_function(output, label2)
    
    loss.backward()
    
    optimizer.step()
    
    if i % 10 ==0:
        print(loss.data)
        
print(time.time()-begin)

#y2가 y1보다 빠르다
```
